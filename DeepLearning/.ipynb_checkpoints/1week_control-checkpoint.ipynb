# 조건문 If
* if 조건 else를 활용
* 비교 연산자 사용
* 같은지 확인하려면 ==
* 같지 않다는 !=
* 크거나 작다는 부등호로 비교 >, <
* 크거나 같다, 작거나 같다의 경우에는 <=, >=

```
money = 2000
if money >= 3000:
  print("택시를 타고 가라")
else:
  print("걸어가라")
```

```
if []:
    print("참")
else:
    print("거짓")
```

* 조건을 위한 연산자
* 둘다 참이어야 참: and
* 둘 중 하나만 참이면 참: or
* 거짓이면 참: not

```
money = 2000
card = True
if money >= 3000 or card:
  print("택시를 타고 가라")
else:
  print("걸어가라")
```

* 활용

```
pocket = ['paper', 'handphone']
card = True
if 'money' in pocket:
  print("택시를 타고가라")
else:
  if card:
    print("택시를 타고가라")
  else:
    print("걸어가라")
```

```
pocket = ['paper', 'cellphone']
card = True
if 'money' in pocket:
  print("택시를 타고가라")
elif card:
  print("택시를 타고가라")
else:
  print("걸어가라")
```

# 반복문 for
* for .. in 을 이용해서 반복 수행
* range()를 통해서 필요한 범위를 지정

```
for i in range(5):
    print('hello')
```

```
for i in range(5):
    print(i)
```

* range(start, end, step) 으로 구성, 한개만 입력시 end로 인식
* step은 증가폭

```
for i in range(1,6):
    print(i)
```

```
for i in range(1,10,2):
    print(i)
```

```
for i in range(5,1,-1):
    print(i)
```

* 활용

```
for num1 in range(2,10):
    print()
    print(num1, '단')
    for num2 in range(1,10):
        print('{} * {} = '.format(num1, num2), num1*num2)
```

# 반복문 While
* while 조건: 문법을 통해 조건에 만족 시 반복 수행
* while문을 강제로 종료시키는 break
* while문의 처음으로 돌아가게 하는 continue

```
treeHit = 0
while treeHit < 10:
  treeHit = treeHit +1
  print("나무를 %d번 찍었습니다." % treeHit)
  if treeHit == 10:
    print("나무 넘어갑니다.")
```

```
coffee = 10
money = 300
while money:
  print("돈을 받았으니 커피를 줍니다.")
  coffee = coffee -1
  print("남은 커피의 양은 %d개입니다." % coffee)
  if coffee == 0:
    print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
    break
```

```
a = 0
while a < 10:
  a = a + 1
  if a % 2 == 0:
    continue
  print(a)
```

# Function
* 반복을 쉽도록하는 함수 정의
* def 함수명(입력변수): 실행 의 구조를 가진다

```
def 직사각형넓이(가로, 세로):
    area = 가로 * 세로
    print(area)
```

```
직사각형넓이(10,20)
```

```
직사각형넓이(20,50)
```

* 활용
* a부터 b까지의 합 계산

```
#함수 미사용
start_num = 1
end_num = 100
sum_ = 0
for i in range(start_num, end_num+1):
    sum_ += i # sum_ = sum_ + i
print(sum_)

start_num = 5
end_num = 200
sum_ = 0
for i in range(start_num, end_num+1):
    sum_ += i # sum_ = sum_ + i
print(sum_)
```

```
#함수 사용
def sum_num(start_num, end_num):
    sum_=0
    for i in range(start_num, end_num+1):
        sum_ += i
    return sum_
sum1 = sum_num(1,100)
sum2 = sum_num(50, 110)

print(sum1)
print(sum2)
```
